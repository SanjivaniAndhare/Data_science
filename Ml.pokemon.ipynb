```python
features=dt.iloc[:,:-1] #......2d 
target= dt.iloc[:,0]    #......1d
```

```python
target
```

```python
features
```

```python
features["generation"].replace("?", np.nan,inplace=True)
features["catch_rate"].replace("?", np.nan,inplace=True)

features["generation"]=features["generation"].astype("float64")
features["catch_rate"]=features["catch_rate"].astype("float64")
```

```python
#simple imputer:
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values =np.nan, strategy ="mean")
features[["generation","catch_rate"]] = si.fit_transform(features[["generation","catch_rate"]])
```

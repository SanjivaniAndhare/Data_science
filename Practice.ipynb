# K-Means:-

```python
from sklearn.cluster import KMeans
```

```python
df.head()
```

```python
plt.scatter(df.Age,df["Fare"])
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[["Age","Fare"]])
y_predicted
```

```python
df["cluster"]=y_predicted
```

```python
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1["Fare"],color="green")
plt.scatter(df2.Age,df2["Fare"],color="red")
plt.scatter(df3.Age,df3["Fare"],color="black")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend()
```

```python
xyz=df["Fare"]
```

```python
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
km= KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[["Age","Fare"]])
y_predicted
```

```python
df["cluster"]=y_predicted
```

```python
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1["Fare"],color="green")
plt.scatter(df2.Age,df2["Fare"],color="red")
plt.scatter(df3.Age,df3["Fare"],color="black")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend()
```

```python
sse=[]
kk=[]
for k in range(1,10):
    km=KMeans(n_clusters=k)
    km.fit(df[["Age","Fare"]])
    sse.append(km.inertia_)
    kk.append(k)
```

```python
sse
```

```python
plt.plot(kk,sse)
```

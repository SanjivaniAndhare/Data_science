```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.load_dataset('iris')
```

```python
df=sns.load_dataset('iris')
df
```

```python
## Pairplot:--Pairplot A is also known as scatter plot
sns.pairplot(df)
```

```python
sns.pairplot(df,kind='kde',hue='species')
plt.show()
```

```python
#displot:--
sns.displot(df['sepal_length'],kde=True)
plt.show()
```

```python
#countplot:--
sns.countplot(df)
plt.show()
```

```python
#Barplot:--
sns.barplot(x='sepal_length',y='species',data=df)
```

```python
sns.barplot(y='sepal_length',x='species',data=df)
```

```python
#boxplot:-- box plot represent a data of fve number summery
sns.boxplot(df)
plt.show()
```

```python
sns.boxplot(x='species',y='petal_length',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')  #v=vertical
```

```python
sns.boxplot(data=df,orient='h')     #h=horizontal
```

```python
#Categorised my data based on some pther categories
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=df)
```

```python
#Violin plot helps us to see both data interms of density estimatipon 

sns.violinplot(x='species',y='petal_length',data=df,palette='rainbow')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
##Matplotlib : Histogram:--
 # in histogram , x axis contains a variable and y axis will be a frequency of that variable.
 # we have a simple data set of blood sugur lavel of different patients. 
 # We will try to no. pof patients by blood range and try to fugure out how many patients are normal diabetic and pre-diabetic
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,89]
plt.hist(blood_sugar,rwidth=0.8)  #by default no. od bins are set to be 10
plt.show()
```

```python
#Bins parameter:--
plt.hist(blood_sugar,rwidth=0.5,bins=4)
```

```python
#Histogram showing normal pre-diabetic and diabetic patients distribution
#1.80-100 :Normal
#2.100-125 :pre diabetic
#3.125-150 : Dibetic

```

```python
plt.xlabel('sugar level')
plt.ylabel('no. of patients')
plt.title('Blood Sugar Chart')
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='green')
```

```python
plt.xlabel('sugar level')
plt.ylabel('no. of patients')
plt.title('Blood Sugar Chart')

blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,89]
blood_sugar_women=[67,95,89,120,131,150,84,69,89,79,120,112,100]
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['men','women'])

plt.legend()
```

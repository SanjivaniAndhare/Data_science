```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
exp_val=[1400,6000,300,410,250]
exp_labels=['house rent','food','phoon\electricity bill','car','other utilities']
plt.pie(exp_val,labels=exp_labels)
```

```python
# show percentage for every pie,we can specify radius size to increase hart size
plt.axis('equal')
plt.pie(exp_val,labels=exp_labels,shadow=True,autopct='%1.2f%%',radius=1)
plt.show()
```

```python
#TO explode pie chart:---
plt.axis('equal')
plt.pie(exp_val,labels=exp_labels,shadow=True,autopct='%1.2f%%',radius=1,explode=[0,0,0,0.1,0.2])
plt.show()

```

```python
exp_val=[1400,4000,3000,4100,2500]
exp_labels=['house rent','food','phoon\electricity bill','car','other utilities']
plt.pie(exp_val,labels=exp_labels,autopct='%0.2f%%',radius=1,counterclock =True,startangle = 90,shadow = True,
        wedgeprops ={'edgecolor':'green','linewidth':3})
plt.legend(loc='upper left')
plt.show()
```

```python
exp_val=[1400,4000,3000,4100,2500]
exp_labels=['house rent','food','phoon\electricity bill','car','other utilities']
plt.pie(exp_val,labels=exp_labels,autopct='%0.1f%%',radius=1.5,counterclock =True,startangle = 50,shadow = True,
        wedgeprops ={'edgecolor':'blue','linewidth':2})
plt.legend(loc='upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
days=[1,2,3,4,5,6]
max_t=[5,60,25,77,48,49]
min_t=[41,42,43,33,38,37]
avg_t=[45,46,47,48,50,52]

plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='average')
plt.plot(days,max_t,label='max')
plt.legend(loc='upper right',shadow=True,fontsize='large')
```

```python
days=[1,2,3,4,5,6]
max_t=[50,41,52,47,48,49]
min_t=[21,30,25,30,45,33]
avg_t=[30,15,33,18,22,28]

plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='average')
plt.plot(days,max_t,label='max')
plt.legend(loc='upper right',shadow=True,fontsize='large')
plt.grid()
```

```python
## Simple bar vchart showing Revenues of major US Tech Companies
company=['Amazon','Meta','Google','NSFT']
revenue=[50,100,150,200]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue,label='Revenues')
plt.xticks(xpos,company)
plt.xlabel('Revenue(Bln)')
plt.title('US technology stock')
plt.legend()
```

```python
## multiple bars showing Revenue and Profit of major US tech companies
profit=[46,60,49,50]
```

```python
plt.xticks(xpos,company)
plt.xlabel('Revenue(Bln)')
plt.xlabel('companies')
plt.title('US technology stock')
plt.bar(xpos,revenue,width=0.4,label='Revenue')
plt.bar(xpos,profit,width=0.4,label='profit')
plt.legend()
plt.show()
```

```python
plt.barh(xpos,revenue,label='Revenues')
plt.yticks(xpos,company)
plt.ylabel('Revenue(Bln)')
plt.title('US technology stock')
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.ylabel('Revenue(Bln)')
plt.ylabel('companies')
plt.title('US technology stock')
plt.barh(xpos-0.2,revenue,label='Revenue')
plt.barh(xpos+0.2,revenue,label='profit')
plt.legend()
plt.show()
```

```python
x=(1,2,3,4,5,6,7,)
y=(50,51,52,48,47,49,46)
plt.xlabel('day')
plt.ylabel('temprature')
plt.title('weather forcasting')
plt.plot(x,y,'g+--')
```

```python
#marker size:--
plt.plot(x,y,color='green',marker='+',linestyle= ' ',markersize=10)
```

```python
#Alpha property to control transparency of line chart
plt.plot(x,y,'g<',alpha=0.4, markersize=10)   # alpha can be specified on a scale 0 to 1
```

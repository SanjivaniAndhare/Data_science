```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('CAR DETAILS FROM CAR DEKHO (1).csv')
df
```

```python
df.head()
```

```python
df.describe()
```

```python
## Pairplot:--Pairplot A is also known as scatter plot
sns.pairplot(df)
```

```python
sns.pairplot(df,kind='kde',hue='fuel')
plt.show()
```

```python
#displot:--
sns.displot(df['fuel'],kde=True)
plt.show()
```

```python
#countplot:--
sns.countplot(df)
plt.show()
```

```python
#Barplot:--
sns.barplot(x='year',y='fuel',data=df)
```

```python
sns.barplot(y='year',x='fuel',data=df)
```

```python
#boxplot:-- box plot represent a data of fve number summery
sns.boxplot(df)
plt.show()
```

```python
sns.boxplot(x='fuel',y='year',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
#Categorised my data based on some pther categories
sns.boxplot(y='fuel',x='year',hue='owner',data=df)
```

```python
#Violin plot helps us to see both data interms of density estimatipon 

sns.violinplot(x='fuel',y='year',data=df,palette='rainbow')
```

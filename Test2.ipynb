## 3.Explain type of Inheritance And perform Single inheritance example.

### Inheritance:-
     1. Single Inheritance 
     
     2. Multiple Imheritance
     
     3. Multilevel Inheritance


```python
#1.Single Inheritance:-
class clg:
    def info1(self,cname,city):
      print('clg name:',cname)
      print('clg city:',city)
class student(clg):
    def info2(self,id,name,year):
      print('student id:',id)
      print('student name:',name)
      print('student year:',year)
stu=student()
stu.info1('Pcfscs','Smbhajinagar')
stu.info2(1,'Achal','3yr')
```
